```python
#import libraries
import numpy as np    #numpy library for algebric calculations
import pandas as pd   #pandas librariy for dataframes dealing
```

```python
from keras.models import Sequential    #import class sequential from package models from keras lib for neural network layers
from keras.layers import Dense         #import dense class to determine numbers of neurals in each layer
```

```python
data = pd.read_csv('forestfires.csv')  #import file
```

# Cleaning Data

```python
#dropping x , y  axis that would not be important about our study
data = data.drop(['X' , 'Y' , 'day'] , axis = 1)
```

```python
data
```

```python
#adding new column that detrmine if the area in the dangerous zone of fires that is > 15
data['danger'] = np.where(data['area'] < 15 , 0 , 1)
```

```python
data
```

```python
#converting string data to neumiric data
data = pd.get_dummies(data)
```

```python
#normalize data
for column in data.columns:
    data[column] = (data[column] - data[column].min()) / (data[column].max() - data[column].min())
```

```python
data
```

```python
data.shape
```

```python
#select the target of study
y = data.loc[ : ,'danger'].values
```

```python
#select features
x = data.drop(['danger' , 'area'] , axis = 1).values
```

```python
x.shape
```

# Neoral Network Model

```python
#making an object from sequential class
Network = Sequential()
```

```python
#make first layer that will take 27 input in 27 neural
Network.add(Dense(20, input_dim=20,  activation='relu'))
#make second layer that consist of 10 neural
Network.add(Dense(10,   activation='relu'))
#make third layer that consist of 10 neural
Network.add(Dense(10,   activation='relu'))
```

```python
#make output neural that consist of 1 neural
Network.add(Dense(1, activation='sigmoid'))
```

```python
#compiling the model
Network.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#train the model with 350 epoch and select 35% of data for testing
training = Network.fit(x, y, validation_split=0.3, epochs=350, batch_size=10)
```

```python
#calculating the accuracy of the model
Accuracy = Network.evaluate(x, y)
print("%s: %.2f%%" % (Network.metrics_names[1], Accuracy[1]*100))
```
